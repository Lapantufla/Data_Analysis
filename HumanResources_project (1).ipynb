````python
# Obtener las variables categoricas y las numericas
num_vars = []
cat_vars = []
for col in df.columns:
  if df[col].dtype == np.object_:
    cat_vars.append(df[col].name)
  else:
    num_vars.append(df[col].name)
  
# [1:-1] para no incluir a la calomna attrition
print(cat_vars)
print(num_vars)
````

Procedemos a tratar las variables categóricas con dos aprouch : **OneHotEncoder** y **OrdinalEncoder**.

````python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
````

````python
one_hot_enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
ord_enc = OrdinalEncoder()

#Removemos de la lista de cat_vars el string Gender ya que esta variable sera tratada con un aprouch distinto
cat_vars_ohe = cat_vars[1:4] + cat_vars[5:-1]
cat_vars_oe = [cat_vars[0],cat_vars[-1],cat_vars[4]]

# Codificamos las variables : 'BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus' con one hot encoder
df_one_hot_enc = pd.DataFrame(one_hot_enc.fit_transform(df[cat_vars_ohe]))

# Seteamos el nombre de las nuevas columnas
df_one_hot_enc.columns = one_hot_enc.get_feature_names_out(cat_vars_ohe)

# Hacemos una copia en las columnas a aplicar el ordinal econder para no perder el nombre de las columnas
df_ord_enc = df[cat_vars_oe].copy()

# Codificamos la variable Gender con ordinal encoder y renombramos la columna
df_ord_enc[cat_vars_oe] = pd.DataFrame(ord_enc.fit_transform(df[cat_vars_oe]))
````

Verificamos la codificacion

````python
for row in range(len(df["OverTime"])):
  #Chequeamos que la columna overtime haya sido codificada bien
  if df_ord_enc['OverTime'][row] == 1:
    if df["OverTime"][row] == "No":
      print("Hubo un error")
  
  if df_ord_enc['OverTime'][row] == 0:
    if df["OverTime"][row] == "Yes":
      print("Hubo un error")

  #Chequeamos que la columna Gender haya sido codificada bien
  if df_ord_enc['Gender'][row] == 1:
    if df["Gender"][row] == "Female":
      print("Hubo un error")
  
  if df_ord_enc['Gender'][row] == 0:
    if df["Gender"][row] == "Male":
      print("Hubo un error")

  #Chequeamos que la columna attrition haya sido codificada bien

  if df_ord_enc['Attrition'][row] == 1:
    if df["Attrition"][row] == "No":
      print("Hubo un error")
  
  if df_ord_enc['Attrition'][row] == 0:
    if df["Attrition"][row] == "Yes":
      print("Hubo un error")
````

Una vez tratadas las variables categóricas pasamos a concatenar las variables que eran categóricas con las variables numéricas.

````python
features = pd.concat([df_one_hot_enc, df_ord_enc, df[num_vars]], join="outer", axis=1)

Y = features.pop("Attrition")
````

## Escalar datos

<font face = "Microsoft Yahei"  color="1E90FF" >El escalado de datos se suele hacer para que al declarar nuestro modelo de ML no infiera resultados a partir de pasar un input donde algunas columnas puedan tener datos más elevados que otras pero no tener ninguna importancia más notoria sobre las columnas donde existan valores más bajos (p ejm columna de StockOptionLevel vs columna de TotalWorkingYears, aquí no necesariamente la columna TotalWorkingYears pudiera tener más importancia que la columna de StockOptionLevel)<font face  />

````python
# Checkeamos que no haya columnas categoricas para poder aplicar el escalado
for col in features.columns:
  if features[col].dtype == np.object_:
    print(col)
````

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler_features = scaler.fit_transform(features)
````

# ENTRENAR Y EVALUAR UN CLASIFICADOR UTILIZANDO REGRESIÓN LOGÍSTICA

## División de datos

````python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaler_features, Y, train_size=0.8)
````

## Declaración y entrenamiento del modelo de regresión logística

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)
````

````python
y_pred = pd.DataFrame(y_pred).rename(columns={0:"Prediction"})
y_pred
````

Podemos ver que nuestro modelo no tiene una precisión muy alta **cuando se trata de predecir los empleados que abandonaran la empresa**. Gracias a la métrica de recall podemos ver que el modelo está pudiendo identificar **solo la mitad** de los empleados que abandonan la empresa.

Esto pudiera ser porque el modelo de regresión logística no se estaría ajustando bien a nuestro problema por lo que optar por otro modelo de clasificación podría ser lo mejor, o tmb podría ser que los features seleccionados con los que entrenamos al modelo no son los más óptimos para la predicción de los empleados que abandonaran la empresa, estos podrían ser algunas explicaciones de porque obtenemos una métrica de recall baja.



````python
from sklearn.metrics import confusion_matrix, classification_report

print(f"Acurracy {100*accuracy_score(y_test, y_pred)}")
print("-------------------------")
print((classification_report(y_test, y_pred)))
````
